```python
import random

class SortingApp:
    def __init__(self):
        self.array = []
        self.is_first_run = True

    def get_user_input(self):
        if not self.is_first_run:
            choice = input("\nDo you want to (1) use the current array, (2) input a new array, (3) generate a random array, or (4) exit? Enter 1, 2, 3, or 4: ").strip()
            if choice == "1":
                return True
            elif choice == "2":
                self.array = list(map(int, input("Enter numbers separated by spaces: ").split()))
            elif choice == "3":
                size = int(input("Enter the size of the random array (max 10): "))
                if size > 10:
                    print("Size exceeds the limit of 10. Setting size to 10.")
                    size = 10
                lower_bound = int(input("Enter the lower bound for random values: "))
                upper_bound = int(input("Enter the upper bound for random values: "))
                self.array = [random.randint(lower_bound, upper_bound) for _ in range(size)]
            elif choice == "4":
                print("Exiting program. Goodbye!")
                return False
            else:
                print("Invalid choice. Please try again.")
                return self.get_user_input()
        else:
            choice = input("Do you want to (1) input an array, (2) generate a random array, or (3) exit? Enter 1, 2, or 3: ").strip()
            if choice == "1":
                self.array = list(map(int, input("Enter numbers separated by spaces: ").split()))
            elif choice == "2":
                size = int(input("Enter the size of the random array (max 10): "))
                if size > 10:
                    print("Size exceeds the limit of 10. Setting size to 10.")
                    size = 10
                lower_bound = int(input("Enter the lower bound for random values: "))
                upper_bound = int(input("Enter the upper bound for random values: "))
                self.array = [random.randint(lower_bound, upper_bound) for _ in range(size)]
            elif choice == "3":
                print("Exiting program. Goodbye!")
                return False
            else:
                print("Invalid choice. Please try again.")
                return self.get_user_input()

        self.is_first_run = False
        return True

    def choose_sorting_algorithm(self):
        print("\nChoose a sorting algorithm:")
        print("1. Bubble Sort")
        print("2. Insertion Sort")
        print("3. Merge Sort")
        print("4. Quick Sort")
        print("5. Selection Sort")
        print("6. Exit")

        choice = input("Enter the number of your choice: ").strip()
        if choice == "1":
            print("Bubble Sort: Repeatedly swaps adjacent elements if they are in the wrong order.")
            return self.bubble_sort
        elif choice == "2":
            print("Insertion Sort: Builds the sorted array one element at a time by inserting elements into their correct position.")
            return self.insertion_sort
        elif choice == "3":
            print("Merge Sort: Divides the array into halves, sorts each half, and merges them back together.")
            return self.merge_sort
        elif choice == "4":
            print("Quick Sort: Picks a pivot element and partitions the array into elements less than and greater than the pivot.")
            return self.quick_sort
        elif choice == "5":
            print("Selection Sort: Selects the minimum element and places it in the correct position.")
            return self.selection_sort
        elif choice == "6":
            return None
        else:
            print("Invalid choice. Please try again.")
            return self.choose_sorting_algorithm()

    def selection_sort(self, array):
        size = len(array)
        for s in range(size):
            min_idx = s
            print(f"Pass {s+1}: {array}")

            for i in range(s + 1, size):
                if array[i] < array[min_idx]:
                    min_idx = i

            if min_idx != s:
                print(f"  Swapping {array[s]} and {array[min_idx]}")
                array[s], array[min_idx] = array[min_idx], array[s]
            print(f"Array after pass {s+1}: {array}")
        return array

    def bubble_sort(self, array):
        n = len(array)
        for i in range(n):
            already_sorted = True
            print(f"Pass {i+1}: {array}")
            for j in range(n - i - 1):
                if array[j] > array[j + 1]:
                    print(f"  Swapping {array[j]} and {array[j + 1]}")
                    array[j], array[j + 1] = array[j + 1], array[j]
                    already_sorted = False
            if already_sorted:
                break
            print(f"Array after pass {i+1}: {array}")
        return array

    def insertion_sort(self, array):
        for i in range(1, len(array)):
            key_item = array[i]
            j = i - 1
            print(f"Inserting {key_item} into sorted part: {array[:i]}")
            while j >= 0 and array[j] > key_item:
                print(f"  Moving {array[j]} to the right")
                array[j + 1] = array[j]
                j -= 1
            array[j + 1] = key_item
            print(f"  Result after inserting {key_item}: {array[:i+1]} + {array[i+1:]}")
        return array

    def merge_sort(self, array):
        if len(array) < 2:
            return array
        mid = len(array) // 2
        print(f"Splitting: {array}")
        left = self.merge_sort(array[:mid])
        right = self.merge_sort(array[mid:])
        merged = self.merge(left, right)
        print(f"Merging: {left} and {right} -> {merged}")
        return merged

    def merge(self, left, right):
        result = []
        index_left = index_right = 0
        while len(result) < len(left) + len(right):
            if index_left < len(left) and (index_right >= len(right) or left[index_left] <= right[index_right]):
                result.append(left[index_left])
                index_left += 1
            else:
                result.append(right[index_right])
                index_right += 1
        return result

    def quick_sort(self, array):
        if len(array) < 2:
            return array
        pivot = array[random.randint(0, len(array) - 1)]
        print(f"Pivot: {pivot}")
        low = [x for x in array if x < pivot]
        same = [x for x in array if x == pivot]
        high = [x for x in array if x > pivot]
        print(f"Low: {low}, Same: {same}, High: {high}")
        return self.quick_sort(low) + same + self.quick_sort(high)

    def run(self):
        if not self.get_user_input():
            return False
        print(f"\nOriginal array: {self.array}")

        original_array = self.array[:]

        sorting_algorithm = self.choose_sorting_algorithm()
        if sorting_algorithm is None:
            print("Exiting program. Goodbye!")
            return False

        sorted_array = sorting_algorithm(original_array)

        print(f"\nSorted array: {sorted_array}")
        return True

if __name__ == "__main__":
    app = SortingApp()
    while True:
        if not app.run():
            break
```
