 <h3> <b> <center> 9. Analyze songs by artist and views-youtube, stream-spotify. </center>  </b> </h3>


```python
def format_number(num):
    if num >= 1e9:
        return f'{num / 1e9:.2f}B'
    elif num >= 1e6:
        return f'{num / 1e6:.2f}M'
    else:
        return str(num)

artist_youtube = data.groupby('Artist')['Views'].sum()
artist_spotify = data.groupby('Artist')['Stream'].sum()

artist_data = pd.DataFrame({'YouTube Views': artist_youtube, 'Spotify Streams': artist_spotify})

artist_data.reset_index(inplace=True)

artist_data = artist_data.sort_values(by='YouTube Views', ascending=False)

artist_data['YouTube Views'] = artist_data['YouTube Views'].apply(format_number)
artist_data['Spotify Streams'] = artist_data['Spotify Streams'].apply(format_number)

print(tabulate(artist_data, headers='keys', tablefmt='mixed_grid', showindex=False))
```

<h3> <b> <center> 10. Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display. </center>  </b> </h3>


```python
# Get the top 10 artists on YouTube and their total stream counts
top_10_youtube_artists = data.groupby('Artist')['Stream'].sum().nlargest(10)
top_10_spotify_artists = data.groupby('Artist')['Views'].sum().nlargest(10)

spotify_streams_youtube = data[data['Artist'].isin(top_10_youtube_artists.index)].groupby('Artist')['Views'].sum()
youtube_streams_spotify = data[data['Artist'].isin(top_10_spotify_artists.index)].groupby('Artist')['Stream'].sum()

combined_streams_youtube = pd.DataFrame({'YouTube Streams': top_10_youtube_artists, 'Spotify Streams': spotify_streams_youtube})
combined_streams_spotify = pd.DataFrame({'Spotify Streams': top_10_spotify_artists, 'YouTube Streams': youtube_streams_spotify})

combined_streams_youtube_sorted = combined_streams_youtube.sort_values(by='YouTube Streams', ascending=False)
combined_streams_spotify_sorted = combined_streams_spotify.sort_values(by='Spotify Streams', ascending=False)

fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot the top 10 artists on YouTube
combined_streams_youtube_sorted.plot(kind='bar', ax=axs[0])
axs[0].set_title('Top 10 Artists on Spotify')
axs[0].set_ylabel('Stream Counts')
axs[0].set_xlabel('Artist')

# Plot the top 10 artists on Spotify
combined_streams_spotify_sorted.plot(kind='bar', ax=axs[1])
axs[1].set_title('Top 10 Artists on Youtube')
axs[1].set_ylabel('Stream Counts')
axs[1].set_xlabel('Artist')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
